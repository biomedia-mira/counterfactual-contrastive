# EDA plots for Fig 4

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib
```

```python
matplotlib.rcParams["font.family"] = "serif"
# Your data
data = {
    "Selenia Dimensions": 0.890791,
    "Senographe Pristina": 0.001744,
    "Senograph 2000D": 0.045019,
    "Lorad Selenia": 0.034735,
    "Clearview CSm": 0.027711,
}

df = pd.Series(data)

fig, ax = plt.subplots(
    ncols=2, nrows=1, figsize=(12, 3), subplot_kw=dict(aspect="equal")
)

wedges, texts = ax[0].pie(
    df,
    wedgeprops=dict(width=0.5),
    startangle=-40,
    colors=plt.get_cmap("Blues")(np.linspace(0.3, 1, 5)),
)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"), zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1) / 2.0 + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = f"angle,angleA=0,angleB={ang}"
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax[0].annotate(
        df.index[i],
        xy=(x, y),
        xytext=((1.2 + i * 0.15) * np.sign(x), (1.5 - (i + 1) * 0.1) * y),
        horizontalalignment=horizontalalignment,
        **kw,
    )


# Your data
data = {
    "Selenia Dimensions": 0.2,
    "Senographe Pristina": 0.2,
    "Senograph 2000D": 0.2,
    "Lorad Selenia": 0.2,
    "Clearview CSm": 0.2,
}

# Create a pandas Series for easier handling
df = pd.Series(data)

wedges, texts = ax[1].pie(
    df,
    wedgeprops=dict(width=0.5),
    startangle=-35,
    colors=plt.get_cmap("Blues")(np.linspace(0.3, 1, 5)),
)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"), zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1) / 2.0 + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = f"angle,angleA=0,angleB={ang}"
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    print(x, y)
    ax[1].annotate(
        df.index[i],
        xy=(x, y),
        xytext=((1.2 + i * 0.15) * np.sign(x), 1.2 * y + i * 0.1),
        horizontalalignment=horizontalalignment,
        **kw,
    )

# Show the plot
ax[0].set_title("Original training set\n(SimCLR, DINO)")
ax[1].set_title(
    "Counterfactual augmented training set\n(SimCLR+, CF-SimCLR, DINO+, CF-DINO)"
)
plt.tight_layout()
plt.show()
fig.savefig("pie_embed.pdf", bbox_inches="tight")
```
